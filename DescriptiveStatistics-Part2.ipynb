## Continuous  Uniform Distribution
- The average weight gained by a person over the winter months is uniformly distributed and ranges from 0 to 30 lbs. 
- Using <b>scipy.stats.uniform()</b> and <b>numpy.random.uniform()</b>
    - Find the probability of a person that he will gain between 10 and 15 lbs in the winter months.
    - Plot the probability distribution in both cases.
    - Calculate the mean and variance,

### Using SciPy

```python
from scipy.stats import uniform
```

```python
wgt = uniform(loc=0, scale=30)
```

```python
wgt.mean()
```

```python
wgt.var()
```

```python
wgt_data = wgt.rvs(size=1)
sns.histplot(wgt_data)
plt.title("a=0 and b=30")
```

```python

# In the standard form, the distribution is uniform on [0, 1]. 
# Using the parameters loc and scale, one obtains the uniform distribution on [loc, loc + scale].

```

```python
prob = wgt.cdf(15) - wgt.cdf(10)
```

```python
prob
```

```python
print(f'Probability of a person that he will gain between 10 and 15 lbs in the winter months:{prob}')
```

### Using NumPy

```python
random_data = np.random.uniform(0, 30, size=10000)
```

```python
random_data.mean()
```

```python
random_data.var()
```

```python
sns.histplot(x=random_data,discrete=True)
plt.title("a=0 and b=30")
```

```python
prob_numpy = np.sum((random_data >= 10) & (random_data <= 15)) / len(random_data)
prob_numpy
```

```python
print(f'Probability of a person that he will gain between 10 and 15 lbs in the winter months:{prob_numpy}')
```
